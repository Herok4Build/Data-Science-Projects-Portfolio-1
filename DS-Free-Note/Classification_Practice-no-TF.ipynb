Using the <code>SelectKBest</code> method with mutual info classifier metric to select the best eight features.
Mutual information is further described within the SKLearn documentation. [SKLearn docs on mutual information.](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.mutual_info_classif.html#sklearn.feature_selection.mutual_info_classif) The notes on mutual information that follow are paraphrased from SKLearn. 
Having a value of 0 is indicative of two random variables having no dependencies on one another. If the value is non-zero, that can be indicative of some dependencies between the two variables in question. The mutual information metric is based on KNN.

```python
from sklearn.feature_selection import mutual_info_classif
def featureSelectionOfKBestResult(submitted_data, submittedLabels, focused_number):
    np.random.seed(90)
    transform_by_features = SelectKBest(score_func=mutual_info_classif,k=focused_number)
    transform_by_features.fit(X=submitted_data, y=submittedLabels)
    print(transform_by_features)
    transformed_data = transform_by_features.fit_transform(submitted_data, submittedLabels) 
    print(transformed_data)
    return transformed_data

transformed_data = featureSelectionOfKBestResult(standardized_generated_dataframe, generated_labels, 8)
```

<h2>Cross-Validation</h2>
<p>Implmenting the cross-validation for the selected model. This will be done for an assortment of models and their resulting performance will be accessed.</p>

```python
def cross_validate_model_function(model,data,labels):
    resultsFromCV = cross_validate(estimator=model, X=data, y=labels,cv = 10,
                                   n_jobs = 3,
                                   #pre_dispatch =1,
                                   return_train_score=True, 
                                   scoring= {"accuracy": "accuracy", 
                                             "balanced_accuracy": "balanced_accuracy",
                                             "F1": "f1_micro"
                                            })
    return resultsFromCV
```

## MLP

<p>We will be using multilayer perceptron neural network below with <code>activation="relu"</code> and <code>solver=adam</code>. The neural network consists of an input layer that accepts the input data, the hidden layers where calculations are made to perform computations on supplied features, then the ouput layer that will produce the resulting prediction. Neural networks are a more complex predictive modeling method, and can be difficult to interpret.</p> 

```python
mlp_neural_network_cross_validation = cross_validate_model_function(
    mlp_neural_network_cl,pd.DataFrame(data=transformed_origin_generated_data),generated_labels)
mlp_neural_network_cross_validation_v2 = cross_validate_model_function(
    mlp_neural_network_cl_v2,pd.DataFrame(data=transformed_origin_generated_data),generated_labels)
```

```python
import numpy as np
def get_model_scores(score_key_list, score_dict):
    for key in score_key_list:
        print(key," (median): " ,round(np.median(score_dict[key]),3))
        print(key, " (mean): ", round(np.mean(score_dict[key]),3), "\n")

def score_extract(score_dict):
    train_score_list = [key for key in score_dict if "train" in key]
    test_score_list = [key for key in score_dict if "test" in key]
    get_model_scores(train_score_list, score_dict)
    get_model_scores(test_score_list, score_dict)
    
print("The training and testing scores for first MLP, learning rate = 0.009: \n")
score_extract(mlp_neural_network_cross_validation)


print("\n The training and testing scores for second MLP, learning rate = 0.01: \n")
score_extract(mlp_neural_network_cross_validation_v2)
```

## MLP Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
# form a pipline for the MLP model
# start with the select K best with mutual information.
base_pipeline_mutual_info =[("mutual_info", SelectKBest(score_func=mutual_info_classif))]
mutual_info_params = {"mutual_info__k": [3, 5, 6, 7, 8, 10]}
# activation="relu", solver="adam", learning_rate_init=0.05,max_iter=100
#MLP_pipeline = base_pipeline_mutual_info.copy()
MLP_pipeline= base_pipeline_mutual_info + [("mlp_nn", MLPClassifier() )]
#MLP_pipeline = Pipeline(MLP_pipeline)
MLP_pipeline_params = mutual_info_params.copy()
MLP_pipeline_params.update({"mlp_nn__hidden_layer_sizes": [(8,3), (6,4), (10,2)],
                                                       "mlp_nn__activation": ["relu", "logistic"],
                                                       "mlp_nn__learning_rate_init": [0.01, 0.05, 0.1],
                                                       "mlp_nn__max_iter": [200,400]})
```

```python
MLP_pipeline
```

```python
MLP_pipeline_params
```

```python
MLP_grid_search = GridSearchCV(estimator = Pipeline(steps = MLP_pipeline), 
                               param_grid = MLP_pipeline_params, 
                               scoring= {"accuracy": "accuracy", 
                                             "balanced_accuracy": "balanced_accuracy",
                                             "F1": "f1_micro"
                                            }, refit = "F1",
                              n_jobs = 12,#5, 
                               cv = 10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(standardized_generated_dataframe, generated_labels, random_state = 34, test_size = 0.3, shuffle = True)
```

## KNN Pipeline

```python
# form a pipline for the KNN model
# start with the select K best with mutual information.
base_pipeline_mutual_info =[("mutual_info", SelectKBest(score_func=mutual_info_classif))]
mutual_info_params = {"mutual_info__k": [3, 5, 6, 7, 8, 10]}

KNN_pipeline= base_pipeline_mutual_info + [("knn", KNeighborsClassifier())]

KNN_pipeline_params = mutual_info_params.copy()
KNN_pipeline_params.update({"knn__n_neighbors": [3,5,7,11,13,15], # Avoid issues with ties
                                                       "knn__weights": ["uniform", "distance"]})
```

```python
KNN_grid_search = GridSearchCV(estimator = Pipeline(steps = KNN_pipeline), 
                               param_grid = KNN_pipeline_params, 
                               scoring= {"accuracy": "accuracy", 
                                             "balanced_accuracy": "balanced_accuracy",
                                             "F1": "f1_micro"
                                            }, refit = "F1",
                              n_jobs = 12,#5, 
                               cv = 10)
```

```python
np.random.seed(12)
KNN_grid_search.fit(X = x_train, y = y_train)
```

```python
# Get the best parameters overall
KNN_grid_search.best_params_
```

```python
# Get the results from the Grid search cross-validation
for metric in logged_metrics:
    print("The overall average ", metric," is: ", round(np.mean(KNN_grid_search.cv_results_[metric]),3))
    print("The overall max ", metric," is: ", round(np.max(KNN_grid_search.cv_results_[metric]),3))
    print("The overall median ", metric," is: ", round(np.median(KNN_grid_search.cv_results_[metric]),3))
    print("-------------------------------------------------")
```

We get the results for the test set:

```python
print(classification_report(y_pred = KNN_grid_search_predictions, y_true = y_test))
```

Overall, the results of the KNN classifier are far more accurate than that of the MLP classifier. The second attempt with the MLP classifier using a larger numebr of iterations faired better overall. 

Generally, the f1-scores for classes 0, 1, and 3 are the highest for the KNN classifier. The precision for class 0 is 0.76. The recall for class 0 is 0.85. Class 2 may be suffering a bit in terms of the f1-score and recall as a result of being the minority class. We can see that overall, it is a good classifier.

Something to note is both the MLP and KNN under mutual information for feature selection are selecting ten features/predictors to use. Only six features are informative and 21 features are redundant within the simulated dataset.
